```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset

test_dataset = pd.read_csv("train_1.csv", nrows=5)  # Load only the first 5 rows
print(test_dataset)


# Display basic statistics of the dataset
print(test_dataset.describe())

# Handle missing values
test_dataset.dropna(subset=['Embarked', 'Name', 'Cabin', 'Sex', 'Ticket', 'Fare'], inplace=True)
test_dataset['Age'].fillna(test_dataset['Age'].mean(), inplace=True)
test_dataset[['SibSp', 'Parch']] = test_dataset[['SibSp', 'Parch']].fillna(0)



# Explore correlations between features
correlation_matrix = test_dataset.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()


# Assuming 'test_dataset' is your DataFrame
sns.pairplot(test_dataset, hue='Parch' , diag_kind='kde')


plt.title('Parents CHildern Spouses abroad the titanic')
plt.show()


#print(test_dataset)
#print(test_dataset.isnull().sum())
#print(train_dataset.info())

#print(test_dataset.info())
#print(test_dataset.head())
#print(test_dataset.tail())
#print(test_dataset.describe())
#print(test_dataset.shape)

```
