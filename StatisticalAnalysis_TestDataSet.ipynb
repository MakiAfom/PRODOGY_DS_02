```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np





# Specify the encoding (e.g., 'latin1' or 'ISO-8859-1')
test_dataset = pd.read_csv("test_1.csv", encoding='latin1')

# Display the first 5 rows to check if the data is loaded correctly
print(test_dataset.head())


# Explore age distribution
sns.histplot(test_dataset['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Explore fare distribution
sns.histplot(test_dataset['Fare'], bins=30, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
