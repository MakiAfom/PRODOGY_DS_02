```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Assuming 'titanic_df' is your DataFrame

test_dataset = pd.read_csv("train_1.csv", nrows=5)  # Load only the first 5 rows
print(test_dataset)

# Explore age distribution
sns.histplot(test_dataset['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Explore fare distribution
sns.histplot(test_dataset['Fare'], bins=30, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
