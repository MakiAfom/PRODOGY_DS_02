```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np





# Specify the encoding (e.g., 'latin1' or 'ISO-8859-1')
train_dataset = pd.read_csv("train_1.csv")

# Display the first 5 rows to check if the data is loaded correctly
print(train_dataset.head())


# Explore age distribution
sns.histplot(train_dataset['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Explore fare distribution
sns.histplot(train_dataset['Fare'], bins=30, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
